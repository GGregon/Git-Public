```python
# Importando bibliotecas
import numpy as np
import math
from matplotlib.cm import get_cmap
from matplotlib.colors import Normalize,LogNorm,LinearSegmentedColormap
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft, fftfreq,fftshift
from matplotlib.gridspec import GridSpec
import matplotlib.gridspec as gridspec
from scipy.signal import spectrogram
```

## Criando o sinal no tempo 

Começamos por definir a quantidade de amostras que serão usadas por unidade de tempo. Esta quantidade é obtida por meio da  `sampling_rate` - comumente expressa em número de amostras por segundo. Portanto, para encontrar a quantidade de amostras necessárias, basta invertermos a `sampling_rate` e multiplicá-la pela duração do sinal, $\tau$.

```python
# Vetor com os intervalos de tempo
sampling_rate = 1000
T = 1/sampling_rate         
t = np.arange(0, 1, T)
```

Definiremos um sinal como sendo a soma de senos da seguinte forma:

$$
x(t) = \sum_{k=1}^{n} A_k \sin(2\pi f_k t)
$$

```python
# Componentes que constituem o sinal
x1 = 7 * np.sin(2 * np.pi * 30  * t)
x2 = 6 * np.sin(2 * np.pi * 50  * t)
x3 = 5 * np.sin(2 * np.pi * 110 * t)
```

> Na definição acima, os senos estão em fase, mas diferem em amplitude e frequência. O sinal resultante será obtido somando as componentes constituintesm

```python
# Definição do sinal como a soma das componentes
x = x1+x2+x3
```

## Exibindo o sinal $x(t)$ no tempo

Construindo um mapa de cores para as componentes de diferentes frequências

```python
# Mapa de cores
colormap   = get_cmap('BuPu')
cmap_slice = LinearSegmentedColormap.from_list('cmap_slice', 
                                               colormap(np.linspace(0.3, 0.6, 256)))    # Limita o mapa de cores
```

```python
# Normaliza (mapeia) o mapa de cores de acordo com as frequências das ondas
freq_componentes = [30,50,110]
norm = Normalize(vmin=min(freq_componentes), vmax=max(freq_componentes))
```

```python
fig = plt.figure(figsize=(12, 6))
gs = gridspec.GridSpec(2, 1, figure=fig)

ax0 = fig.add_subplot(gs[0])
ax1 = fig.add_subplot(gs[1],sharex=ax0,sharey=ax0)

ax0.plot(t,x,c='tab:blue')
ax0.set_title('Sinal resultante')

color_f1 = cmap_slice(norm(freq_componentes[0]))
color_f2 = cmap_slice(norm(freq_componentes[1]))
color_f3 = cmap_slice(norm(freq_componentes[2]))

ax1.plot(t,x1, label=r'$7 \sin(2\pi \cdot 30 t)$',c=color_f1,lw=1)
ax1.plot(t,x2, label=r'$2 \sin(2\pi \cdot 50 t)$',c=color_f2,lw=1)
ax1.plot(t,x3, label=r'$1 \sin(2\pi \cdot 110 t)$',c=color_f3,lw=1)
ax1.set_yticklabels([])
ax1.set_title('Componentes')
ax1.set_xlabel('Tempo [s]')

ax1.legend()

for ax in [ax0,ax1]:
    ax.set_ylabel('Amplitude')
    ax.minorticks_on()
    ax.grid(True,alpha=0.5,axis='x',which='both',ls='dotted')
    ax.grid(True,alpha=0.5,axis='y',which='major',ls='dotted')

plt.tight_layout()
plt.show()
```

## Analisando o sinal na frequência

Para $x(t) = A \sin(\omega_0 t + \phi)$, sua transformada de Fourier $\mathcal{F}$ é dada por:

$$
X(j\omega) = -j\pi [\delta(\omega-\omega_0) - \delta(\omega+\omega_0)]
$$

Isto é, para o módulo, esperamos dois impulsos nas posições $\pm \omega_0$

```python
# Compute the Fourier Transform of the signal
fft_signal = np.fft.fft(x)
fft_freq = np.fft.fftfreq(len(x), T)
```

```python
# Compute magnitude and phase
magnitude = np.abs(fft_signal)
phase = np.arctan(fft_signal.imag/fft_signal.real)*180/np.pi
```

```python
fig = plt.figure(figsize=(12, 6))
gs = gridspec.GridSpec(2, 1, figure=fig)

ax0 = fig.add_subplot(gs[0])
ax1 = fig.add_subplot(gs[1])


ax0.plot(fft_freq[:len(fft_freq)//2], magnitude[:len(magnitude)//2],color='tab:red')
ax0.set_title(r'Módulo $|X(j\omega)|$',loc='left')
ax0.set_xlabel('Frequencia [Hz]')
ax0.set_ylabel('Magnitude')

ax1.plot(fft_freq[:len(fft_freq)//2], phase[:len(phase)//2],color='tab:red',alpha=0.5)
ax1.set_title(r'Fase $\angle X(j\omega)$',loc='left')
ax1.set_xlabel('Frequencia [Hz]')
ax1.set_ylabel('Fase [graus]')

for ax in [ax0,ax1]:
    ax.set_ylabel('Amplitude')
    ax.minorticks_on()
    ax.grid(True,alpha=0.5,axis='x',which='both',ls='dotted')
    ax.grid(True,alpha=0.5,axis='y',which='major',ls='dotted')

plt.tight_layout()
plt.show()
```

### Espectro de Densidade de Energia

```python
fig = plt.figure(figsize=(12, 3))

ax0 = plt.subplot()
ax0.plot(fft_freq[:len(fft_freq)//2], magnitude[:len(magnitude)//2]**2,color='tab:red')
ax0.set_title(r'$\varepsilon(\omega) = |X(j\omega)|^2 $',loc='left')
ax0.set_xlabel('Frequencia [Hz]')

ax0.minorticks_on()
ax0.grid(True,alpha=0.5,axis='x',which='both',ls='dotted')
ax0.grid(True,alpha=0.5,axis='y',which='major',ls='dotted')

plt.show()
```

```python
# Compute the spectrogram
frequencies, times, Sxx = spectrogram(x, fs=sampling_rate)
```

```python
# Plot the spectrogram
plt.figure(figsize=(12,4))
plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='auto',cmap='Reds')
plt.title('Espectrograma')
plt.xlabel('Tempo [s]')
plt.ylabel('Frequencia [Hz]')
plt.colorbar(label='Intensidade [dB]')
```

> Repare que a região mais escura está justamente nos pontos de ordenadas iguais as frequências constituintes do sinal. Já os pontos mais claros representam pontos onde há interferência destrutiva
