## Distribuição de Cauchy
Também conhecida como distribuição de Lorentz, é uma distribuição de probabilidade contínua. Sua função de densidade de probabilidade (PDF) é dada pela seguinte expressão:

$$
f(x) = \frac{1}{\pi \gamma \left[1 + \left(\frac{x - x_0}{\gamma}\right)^2\right]}
$$

Para $-\infty < x < \infty$

onde:
- $ x $ é a variável aleatória.
- $ x_0 $ é a localização do pico da distribuição (parâmetro de localização).
- $ \gamma $ é o parâmetro de escala, que controla a largura da distribuição.

> Essa distribuição é simétrica em torno de $ x_0 $, mas não tem média ou variância finitas devido a suas caudas pesadas.

### Cálculo dos primeiros momentos

### Primeiro momento $M_{X}^{(1)}(0)=\mathsf{E} X$

$$
E |X| = \int_{-\infty}^{\infty} |x|  \frac{1}{\pi \gamma \left[1 + \left(\frac{x - x_0}{\gamma}\right)^2\right]} \, dx
$$

Como estamos integrando de menos infinito a mais infinito, a convergência (ou não) desta integral não será alterada pelos valores dos parâmetros $\gamma$ e $x_0$ escolhidos. A fim de facilitar a "cara" desta integral, fazemos:

$\gamma = 1$ e $x_0=0$

$$
E |X| = \int_{-\infty}^{\infty} \frac{|x|}{\pi}  \frac{1}{1 + x^2}dx =  \frac{2}{\pi} \int_{0}^{\infty} \frac{x}{1 + x^2}dx
$$

> $f(x)$ é simétrica em torno de $x_0=0$

Para resolver essa integral, podemos fazer uma substituição $ u = 1 + x^2 $, então $ du = 2x \, dx $, o que nos leva a $ x \, dx = \frac{1}{2} du $. Com essa substituição, a integral se torna:

$$
\frac{1}{2} \int_{1}^{1 + M^2} \frac{1}{u} \, du
$$

Portanto, 

$$
E |X| = \frac{1}{\pi} \lim_{M\rightarrow \infty} \ln(1 + M^2) = \infty
$$

> De posse do resultado anterior, é evidente que $\text{Var}X = \mathsf{E} X^2 - (\mathsf{E}X)^2$ segue indeterminada

Para visualizar melhor a hipótese de que analisar um caso particular $x_0, \gamma = (0,1)$ não alteraria a convergência da função.

```python
from scipy.stats import cauchy

def cauchy_distribution(x, x0, gamma):
    return cauchy.pdf(x, loc=x0, scale=gamma)
```

```python
# Valores de x para os quais calcular a densidade de probabilidade
x_values = np.linspace(0, 10, 1000)

# Cria grade de gráficos e a figura
gs = GridSpec(1, 2, width_ratios=([1, 1]))
plt.figure(figsize=(12, 3), dpi=200)

ax1 = plt.subplot(gs[0])
ax2 = plt.subplot(gs[1])

# Mapa de cores para distribuição de Cauchy com x0 variável
mapa_x0 = plt.cm.ScalarMappable(cmap='Purples')
mapa_gamma = plt.cm.ScalarMappable(cmap='Blues')

offsets = np.linspace(1,10,10) 
gammas  = np.linspace(1,10,10) 

for i in offsets:
    y_values = cauchy_distribution(x_values, i, 1)
    cor_linha = mapa_x0.to_rgba(i/10)
    ax1.plot(x_values, y_values, color=plt.cm.Purples(i/10))
    

for j in gammas:
    y_values = cauchy_distribution(x_values, 0, j)
    cor_linha = mapa_gamma.to_rgba(j/10)
    ax2.plot(x_values, y_values, color=plt.cm.Blues(j/10))
    
ax1.set_xlabel(r'$X$')
ax1.grid(True, alpha=0.2)

ax2.set_xlabel(r'$X$')
ax2.grid(True, alpha=0.2)
    
cbar = plt.colorbar(plt.cm.ScalarMappable(cmap='Purples'), ax=ax1, label=r'$x_0$')  
cbar.set_ticks([x0/10 for x0 in offsets])
cbar.set_ticklabels([f'{x0}' for x0 in offsets])
cbar.ax.yaxis.label.set_rotation(0)

cbar = plt.colorbar(plt.cm.ScalarMappable(cmap='Blues'), ax=ax2, label=r'$\gamma$')  
cbar.set_ticks([gmm/10 for gmm in gammas])
cbar.set_ticklabels([f'{gmm}' for gmm in gammas])
cbar.ax.yaxis.label.set_rotation(0)

plt.suptitle(r'Distribuição de probabilidades para $X \text{~Cauchy}(x_0, \gamma)$')
# plt.tight_layout()
plt.show()
```

# Análise de Dados Financeiros com Distribuição de Cauchy
Neste notebook, vamos baixar dados financeiros históricos para o índice S&P 500 e analisar sua distribuição de retornos diários. Vamos ajustar uma distribuição de Cauchy aos dados e visualizar tanto os dados empíricos quanto a distribuição de Cauchy ajustada.

## 1. Sobre o conjunto de dados
O Índice S&P 500, ou Standard & Poor's 500, é um índice de mercado de ações que representa o desempenho de 500 das maiores empresas publicamente negociadas nos Estados Unidos. Ele é amplamente considerado um dos melhores indicadores do mercado de ações dos EUA e é amplamente seguido por investidores e analistas financeiros como uma medida do desempenho geral do mercado de ações. O S&P 500 abrange empresas de diversos setores, incluindo tecnologia, saúde, finanças, energia e muito mais, proporcionando uma visão abrangente do mercado de ações dos EUA.

## 2. Importando pacotes
Vamos começar importando as bibliotecas necessárias para análise de dados e visualização.

```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib import colormaps
from scipy.stats import cauchy
```

```python
# LaTeX
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "latinmodern-math",
    "text.latex.preamble": r"\usepackage{amsmath}",
})
```

## 3. Obtenção dos dados
Vamos baixar os dados históricos de retornos diários para o índice S&P 500 usando a API do Yahoo Finance.

```python
# Define o símbolo do ativo e o intervalo de datas
ativo = "^GSPC"  # Índice S&P 500
data_inicial = "2020-01-01"
data_final = "2022-01-01"

# Baixa os retornos diários históricos do Índice S&P 500
dados = yf.download(ativo, start=data_inicial, end=data_final)
retornos = dados["Adj Close"].pct_change().dropna()  # Calcula os retornos diários
```

## 4. Explorando o conjunto de dados
A etapa anterior inicializou a variável `dados` como uma estrutura do tipo `pd.dataframe`.

<b>Open</b><br>
O preço de abertura da ação ou índice no início do período de tempo considerado (por exemplo, o preço no início do dia de negociação).

<b>High</b><br>
O preço mais alto alcançado pela ação ou índice durante o período de tempo considerado (por exemplo, o preço mais alto durante o dia de negociação).

<b>Low</b><br>
O preço mais baixo alcançado pela ação ou índice durante o período de tempo considerado (por exemplo, o preço mais baixo durante o dia de negociação).

<b>Close</b><br>
O preço de fechamento da ação ou índice no final do período de tempo considerado (por exemplo, o preço no final do dia de negociação).

<b>Adj Close (Adjusted Close)</b><br>
O preço de fechamento ajustado da ação ou índice, que leva em consideração eventuais eventos corporativos, como divisões de ações, dividendos, etc. O preço ajustado é útil para análises de longo prazo, pois reflete o preço de fechamento que leva em conta eventos que podem afetar o valor total da ação.

<b>Volume</b><br>
O volume de negociação, que representa o número total de ações ou contratos negociados durante o período de tempo considerado. O volume é uma medida da atividade de negociação e é útil para avaliar a liquidez de um ativo.

## Plotando os Dados Empíricos
Vamos plotar um histograma dos retornos diários para visualizar a distribuição dos dados.

```python
plt.figure(figsize=(10, 5))
plt.hist(retornos, bins=50, density=True, alpha=0.6, color='g', label='Empirical Data')
plt.show()
```

## Ajustando a Distribuição de Cauchy
Vamos ajustar uma distribuição de Cauchy aos dados empíricos e plotá-la ao lado do histograma.

```python
# Fit Cauchy distribution to the data
params = cauchy.fit(retornos)
x = np.linspace(retornos.min(), retornos.max(), 100)
pdf_fitted = cauchy.pdf(x, *params)

# Plot Cauchy distribution
plt.plot(x, pdf_fitted, 'r-', lw=2, label='Cauchy Distribution')
plt.title('Cauchy Distribution Fit to S&P 500 Returns', loc='left')  # Adjust title location
plt.xlabel('Daily Returns')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
