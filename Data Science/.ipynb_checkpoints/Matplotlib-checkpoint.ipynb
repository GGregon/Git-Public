# Introdução ao pyplot

`matplotlib.pyplot` é uma coleção de funções que faz o matplotlib funcionar como o MATLAB. Cada função do `pyplot` faz alguma alteração em uma figura: por exemplo, cria uma figura, cria uma área de plotagem em uma figura, traça algumas linhas em uma área de plotagem, decora o gráfico com rótulos, etc.

No `matplotlib.pyplot`, vários estados são preservados através de chamadas de função, de modo que ele mantém o controle de coisas como a figura e a área de plotagem atuais, e as funções de plotagem são direcionadas para os eixos atuais (observe que "axes" aqui e na maioria dos lugares na documentação se refere à parte de eixos de uma figura e não ao termo matemático estrito para mais de um eixo).

Primeira visualização

````python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
````

> Você pode estar se perguntando por que o eixo $x$ varia de 0-3 e o eixo $y$ de 1-4. Se você fornecer uma única lista ou matriz para `plot`, o matplotlib assume que é uma sequência de valores $y$ e gera automaticamente os valores $x$ para você. 

> Como os intervalos em Python começam com 0, o vetor $x$ padrão tem o mesmo comprimento que $y$, mas começa com 0; portanto, os dados $x$ são `[0, 1, 2, 3]`

A função `plot` é versátil e aceita um número arbitrário de argumentos. Por exemplo, para plotar $x$ versus $y$, você pode escrever:

````python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
````

## Formatando o estilo do seu gráfico
Para cada par de argumentos $x$, $y$, há um terceiro argumento opcional que é a string de formato que indica a cor e o tipo de linha do gráfico. As letras e símbolos da string de formato são do MATLAB, e você concatena uma string de cor com uma string de estilo de linha. A string de formato padrão é 'b-', que é uma linha azul sólida. Por exemplo, para plotar o acima com círculos vermelhos, você usaria:

````python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis((0, 6, 0, 20))
plt.show()
````

Consulte a documentação do `plot` para obter uma lista completa de estilos de linha e strings de formato. A função <b>`axis`</b> no exemplo acima recebe uma lista de `[xmin, xmax, ymin, ymax]` e especifica a janela de visualização dos eixos.


Se o matplotlib estivesse limitado a trabalhar com listas, seria bastante inútil para processamento numérico. Geralmente, você usará arrays do <b>`numpy`</b>. Na verdade, todas as sequências são convertidas em arrays do numpy internamente. O exemplo abaixo ilustra o traçado de várias linhas com diferentes estilos de formato em uma chamada de função usando arrays.

````python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
````

## Traçando com strings de palavras-chave

Existem alguns casos em que você tem dados em um formato que permite acessar variáveis específicas com strings. Por exemplo, com arrays estruturados ou <b>`pandas.DataFrame`</b>.

O Matplotlib permite que você forneça tal objeto com o argumento de palavra-chave data. Se fornecido, você pode gerar gráficos com as strings correspondentes a essas variáveis.

````python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
````

## Traçando com variáveis categóricas
Também é possível criar um gráfico usando variáveis categóricas. O Matplotlib permite que você passe variáveis categóricas diretamente para muitas funções de traçado. Por exemplo:

````python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
````

## Controlando as propriedades das linhas

As linhas possuem muitos atributos que você pode configurar: largura da linha, estilo de traço, anti-aliasing, etc; consulte matplotlib.lines.Line2D. Existem várias maneiras de configurar as propriedades das linhas:

* Utilizando argumentos de palavra-chave (<i>keyword arguments</i>)

```python
plt.plot(x, y, linewidth=2.0)
```

* Utilize os métodos setter de uma instância <b>`Line2D`</b>. A função <b>`plot`</b> retorna uma lista de objetos <b>`Line2D`</b>; por exemplo, <b>`line1`</b>, <b>`line2 = plot(x1, y1, x2, y2)`</b>. No código abaixo, vamos supor que temos apenas uma linha para que a lista retornada tenha comprimento 1. Utilizamos desempacotamento de tupla com <b>`line`</b>, para obter o primeiro elemento dessa lista:

```python
line, = plt.plot(x, y, '-')
line.set_antialiased(False) # turn off antialiasing
```

````python
lines = plt.plot([1, 2, 3])
plt.setp(lines)
````

## Trabalhando com múltiplas figuras e eixos
O MATLAB, e o <b>`pyplot`</b>, têm o conceito de figura atual e eixos atuais. Todas as funções de traçado se aplicam aos eixos atuais. A função `gca` retorna os eixos atuais (uma instância `matplotlib.axes.Axes`), e `gcf` retorna a figura atual (uma instância `matplotlib.figure.Figure`). Normalmente, você não precisa se preocupar com isso, pois tudo é cuidado nos bastidores. Abaixo está um script para criar dois subplots.

````python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
````

<b>``</b>

A chamada <b>`figure`</b> aqui é opcional porque uma figura será criada se nenhuma existir, assim como um Axes será criado (equivalente a uma chamada explícita de `subplot()`) se nenhum existir. A chamada <b>`subplot`</b> especifica <b>`numrows`</b>, <b>`numcols`</b>, <b>`plot_number`</b>, onde o <b>`plot_number`</b> varia de 1 a `numrows*numcols`. As vírgulas na chamada de subplot são opcionais se `numrows*numcols<10`. Então `subplot(211)` é idêntico a `subplot(2, 1, 1)`.

Você pode criar um número arbitrário de subplots e eixos. Se você quiser colocar um Axes manualmente, ou seja, não em uma grade retangular, use <b>`axes`</b>, que permite especificar a localização como axes([esquerda, inferior, largura, altura]) onde todos os valores estão em coordenadas fracionárias (de 0 a 1). Veja o demo de Axes para um exemplo de colocação manual de eixos e múltiplos subplots para um exemplo com muitos subplots.

Você pode criar várias figuras usando múltiplas chamadas de `figure` com um número de figura crescente. Claro, cada figura pode conter quantos eixos e subplots desejar:

````python
plt.figure(1)                # the first figure

plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])

plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot() by default

plt.figure(1)                # first figure current;
                             # subplot(212) still current
plt.subplot(211)             # make subplot(211) in the first figure
                             # current
plt.title('Easy as 1, 2, 3') # subplot 211 title
````

Você pode limpar a figura atual com `clf` e os eixos atuais com `cla`. Se você achar irritante que os estados (especificamente a imagem atual, figura e eixos) sejam mantidos para você nos bastidores, não se desespere: isso é apenas uma fina camada de envoltório stateful em torno de uma API orientada a objetos, que você pode usar em vez disso (consulte o Artist tutorial).

Se você estiver criando muitas figuras, precisa estar ciente de mais uma coisa: a memória necessária para uma figura não é completamente liberada até que a figura seja explicitamente fechada com `close`. Excluir todas as referências à figura e/ou usar o gerenciador de janelas para fechar a janela na qual a figura aparece na tela não é suficiente, porque pyplot mantém referências internas até que `close` seja chamado.
