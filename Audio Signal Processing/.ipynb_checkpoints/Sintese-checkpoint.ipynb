```python
# Manipulação dos dados e plotagem
import numpy as np
import matplotlib.pylab as plt

# Processamento de audio
import librosa
import librosa.display
import IPython.display as ipd
```

## Lê o arquivo de audio

```python
audio_files = r'.\Audio\grave.wav'
```

```python
# Exibe reprodutor de audio
ipd.Audio(audio_files)
```

```python
y, sr = librosa.load(path=audio_files)
```

```python
print(f'Número de amostras            : {y.size}')
print(f'Frequência de amostragem (sr) : {sr} amostras/segundo')
print(f'Duração do sinal (audio)      : {round(y.size/sr,2)} segundos')
```

> O ouvido humano é capaz de ouvir sons de máxima frequência em ~40KHz. A frequência de amostragem ($f_s$) usada é ligeiramente maior que metade desta frequência, conforme o teorema de **Nyquist**.

```python
# Obtendo os instantes de tempo a partir da frequencia de amostragem
t = np.linspace(0,y.size/sr,y.size)
```

```python
plt.figure(figsize=(16,4),dpi=200)
plt.plot(t,y)
plt.grid(alpha=0.3,ls='--')
plt.minorticks_on()
plt.tick_params(axis='both',labelsize=16)
plt.xlabel('Tempo (s)',loc='right',fontsize=16)
plt.title('$\Delta p$',loc='left',fontsize=16)
```

> Note que há regiões de silêncio (amplitudes muito baixas) que podem ser ignoradas.

```python
# Aparando as regiões de silêncio
y_trimmed, trimmed_idx = librosa.effects.trim(y, top_db=20)
```

```python
t_trimmed = t[trimmed_idx[0]:trimmed_idx[1]]
t_trimmed = t_trimmed - min(t_trimmed)
```

```python
plt.figure(figsize=(16,4),dpi=200)
plt.plot(t_trimmed,y_trimmed)
plt.grid(alpha=0.3,ls='--')
plt.minorticks_on()
plt.tick_params(axis='both',labelsize=16)
plt.xlabel('Tempo (s)',loc='right',fontsize=16)
plt.title('$\Delta p$',loc='left',fontsize=16)
```

Zoom em determinada região

```python
plt.figure(figsize=(16,4),dpi=200)
plt.plot(t_trimmed,y_trimmed)
plt.grid(alpha=0.3,ls='--')
plt.xlim(0.25,0.35)
plt.minorticks_on()
plt.tick_params(axis='both',labelsize=16)
plt.xlabel('Tempo (s)',loc='right',fontsize=16)
plt.title('$\Delta p$',loc='left',fontsize=16)
plt.show()
```

## Transformada de Fourier

```python
from scipy.fft import fft, ifft,fftfreq
```

```python
# Number of sample points
N = y_trimmed.size
# sample spacing
T = 1.0/sr

vogal_fourier = fft(y_trimmed)
freq_fourier  = fftfreq(N, T)
```

```python
# Partes real e imaginaria do sinal
real  = vogal_fourier.real
imag  = vogal_fourier.imag
```

```python
# Magnitude (módulo) e fase da transformada
mag   = abs(vogal_fourier)
phase = np.arctan(imag/real)
```

```python
# Faixa de frequências percebidas pelo ser humano
fmin = 20
fmax = 2e3
```

```python
fig = plt.figure(figsize=(16,8),dpi=200)
gs = plt.GridSpec(nrows=2,ncols=2,figure=fig,width_ratios=[10,1],wspace=0.01)

ax1 = fig.add_subplot(gs[0,:])
ax2 = fig.add_subplot(gs[1,0],sharex=ax1)
ax3 = fig.add_subplot(gs[1,1])

# Magnitude
ax1.fill_between(freq_fourier[:N//2],mag[:N//2],color='tab:blue',alpha=0.5)
ax1.set_xlim(fmin,fmax)
ax1.set_ylabel(r'Modulo $|X(\omega)|$')
ax1.set_xlabel('Frequência (Hz)')

# Fase
ax2.plot(freq_fourier[:N//2],phase[:N//2]*(180/np.pi),lw=0.5)
ax2.set_ylabel(r'Fase $\varphi\{X(\omega)\}$')
ax3.hist(phase[:N//2]*(180/np.pi),orientation='horizontal',histtype='step')
ax3.hist(phase[:N//2]*(180/np.pi),orientation='horizontal',color='tab:blue',alpha=0.2)
ax3.set_yticks([])
ax3.axis('off')
plt.show()
```

> De acordo com o histograma, a fase da transformada tem distribuição similar a uma distribuição uniforme. Nesse caso, podemos considerar que todas as componentes serão igualmente desviadas.

## Espectrograma

```python
np.log10(y_trimmed)
```

```python
fig, ax = plt.subplots()

Pxx, freqs, bins, im = ax.specgram(y_trimmed, NFFT=1024, Fs=sr)

# The `specgram` method returns 4 objects. They are:
# - Pxx: the periodogram
# - freqs: the frequency vector
# - bins: the centers of the time bins
# - im: the .image.AxesImage instance representing the data in the plot
ax.set_xlabel('Time (s)')
ax.set_ylabel('Frequency (Hz)')
fig.colorbar(im, ax=ax,label='Intensidade dB')
plt.show()
```

```python
D = librosa.stft(y)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
S_db.shape
```

```python
# Plot the transformed audio data
fig, ax = plt.subplots(figsize=(10, 5))
img = librosa.display.specshow(S_db,
                              x_axis='time',
                              y_axis='log',
                              ax=ax)
ax.set_title('Spectogram Example', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```

## Mel Spectogram

```python
S = librosa.feature.melspectrogram(y=y,
                                   sr=sr,
                                   n_mels=128 * 2,)
S_db_mel = librosa.amplitude_to_db(S, ref=np.max)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
# Plot the mel spectogram
img = librosa.display.specshow(S_db_mel,
                              x_axis='time',
                              y_axis='log',
                              ax=ax)
ax.set_title('Mel Spectogram Example', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```
