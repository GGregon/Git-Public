# Introduction to Colormaps

A colormap in Matplotlib is a mapping from data values to colors. Colormaps are used in various types of plots like scatter plots, contour plots, and heatmaps to represent data points' magnitude or density.

## Types of Colormaps
Colormaps can be broadly categorized into:

* <b>Sequential Colormaps</b>: <br>
    Used for data that has ordering from low to high (e.g., temperature, height). <br>
    <i>Examples</i>: viridis, plasma, inferno, magma.

* <b>Diverging Colormaps</b>: <br>
    Used for data that has a critical midpoint and diverges towards two extremes (e.g., temperature anomalies).  <br>
    <i>Examples:</i> PiYG, PRGn, coolwarm, bwr.

* <b>Qualitative Colormaps</b>: <br>
Used for data that does not have ordering and is categorical.  <br>
<i>Examples:</i> tab10, tab20, Set3.

* <b>Cyclic Colormaps</b>: <br>
Used for data that is cyclic (e.g., wind direction, hue).  <br>
<i>Examples</i>: hsv, twilight.

## Basic Usage of Colormaps
Let's go through some basic examples of how to use colormaps in different types of plots.

#### Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)

# Create scatter plot
plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()  # Show color scale
plt.show()
```

#### Heat Map

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.rand(10, 10)

# Create heatmap
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.colorbar()  # Show color scale
plt.show()
```

#### Countour Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) ** 10 + np.cos(10 + Y * X) * np.cos(X)

# Create contour plot
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()  # Show color scale
plt.show()
```

## Customizing Colormaps
You can customize colormaps in various ways:

### 1. Reverse a Colormap
To reverse a colormap, append _r to its name:

```python
# Sample data
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)

# Create scatter plot
plt.scatter(x, y, c=colors, cmap='viridis_r')
plt.colorbar()  # Show color scale
plt.show()
```

### 2. Create a Custom Colormap
You can create a custom colormap using `LinearSegmentedColormap` or `ListedColormap`.

```python
from matplotlib.colors import ListedColormap

# Custom colormap
cmap = ListedColormap(['red', 'green', 'blue'])

plt.scatter(x, y, c=colors, cmap=cmap)
plt.colorbar()
plt.show()
```

### 3. Modify Colormap Range
You can normalize the colormap to fit a specific range of data using Normalize.

```python
from matplotlib.colors import Normalize

norm = Normalize(vmin=0, vmax=1)
plt.scatter(x, y, c=colors, cmap='viridis', norm=norm)
plt.colorbar()
plt.show()
```

## Advanced Usage

### 1. Using Colormap with Subplots

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
cmap = plt.get_cmap('viridis')

# First subplot
sc1 = axes[0].scatter(x, y, c=colors, cmap=cmap)
fig.colorbar(sc1, ax=axes[0])

# Second subplot
sc2 = axes[1].scatter(x, y, c=colors, cmap=cmap)
fig.colorbar(sc2, ax=axes[1])

plt.show()
```

### 2. Customizing Colorbar
You can customize the colorbar to improve readability and appearance.

```python
fig, ax = plt.subplots()
sc = ax.scatter(x, y, c=colors, cmap='viridis')

# Customize colorbar
cbar = plt.colorbar(sc)
cbar.set_label('Color Scale')
cbar.set_ticks([0, 0.25, 0.5, 0.75, 1])
cbar.set_ticklabels(['Low', 'Medium-Low', 'Medium', 'Medium-High', 'High'])

plt.show()
```
